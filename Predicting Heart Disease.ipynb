# Data Visualization

### Visualize no. of patients having a heart disease and not having a heart disease

# Logistic Regrassion

```python
#Logistic Regrassion

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix , accuracy_score

X= data.drop(columns=["target"])
y= data["target"]
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3,random_state=42)
logistic_model= LogisticRegression()
logistic_model.fit(X_train,y_train)
y_pred= logistic_model.predict(X_test)

#confusion Matrix
conf_matrix=confusion_matrix(y_test,y_pred)
print("Confusion Matrix")
print(conf_matrix)

#accuracy Score
accuracy =accuracy_score(y_test,y_pred)
print("Accuracy Score :", accuracy)

```

# Decision Tree

```python
# Decision Tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import export_graphviz
import graphviz
import matplotlib.pyplot as plt

X = data.drop(columns=['target'])  # Features
y = data['target']  # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(X_train, y_train)
y_pred = decision_tree_model.predict(X_test)

# confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```

# Random Forest

```python
# Random Forest

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import export_graphviz
import graphviz
import matplotlib.pyplot as plt

X = data.drop(columns=['target'])  
y = data['target']  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
random_forest_model = RandomForestClassifier()
random_forest_model.fit(X_train, y_train)
y_pred = random_forest_model.predict(X_test)

# confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

# accuracy score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)

```

# Comparison Between Models

```python
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score

# Initialize lists to store metrics for each model
conf_matrices = []
class_reports = []
accuracies = []
precisions = []
recalls = []
f1_scores = []

# List of classifiers
classifiers = [
    ("Logistic Regression", logistic_model),
    ("Decision Tree", decision_tree_model),
    ("Random Forest", random_forest_model)
]

# Iterate over each classifier
for clf_name, clf in classifiers:
    # Predict on the test set
    y_pred = clf.predict(X_test)

    # Calculate confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    conf_matrices.append((clf_name, conf_matrix))

    # Calculate classification report
    class_report = classification_report(y_test, y_pred)
    class_reports.append((clf_name, class_report))

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append((clf_name, accuracy))

    # Calculate precision
    precision = precision_score(y_test, y_pred)
    precisions.append((clf_name, precision))

    # Calculate recall
    recall = recall_score(y_test, y_pred)
    recalls.append((clf_name, recall))

    # Calculate F1 score
    f1 = f1_score(y_test, y_pred)
    f1_scores.append((clf_name, f1))

    # Plot confusion matrix as a heatmap
    plt.figure(figsize=(4,3))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="summer",
                xticklabels=['No Disease', 'Disease'],
                yticklabels=['No Disease', 'Disease'])
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title(f'Confusion Matrix - {clf_name}')
    plt.show()

# Print confusion matrices
for clf_name, conf_matrix in conf_matrices:
    print(f"Confusion Matrix - {clf_name}:")
    print(conf_matrix)
    print()

# Print classification reports
for clf_name, class_report in class_reports:
    print(f"Classification Report - {clf_name}:")
    print(class_report)
    print()

# Print accuracy scores
print("Accuracy Scores:")
for clf_name, accuracy in accuracies:
    print(f"{clf_name}: {accuracy:.4f}")

# Print precision scores
print("Precision Scores:")
for clf_name, precision in precisions:
    print(f"{clf_name}: {precision:.4f}")

# Print recall scores
print("Recall Scores:")
for clf_name, recall in recalls:
    print(f"{clf_name}: {recall:.4f}")

# Print F1 scores
print("F1 Scores:")
for clf_name, f1 in f1_scores:
    print(f"{clf_name}: {f1:.4f}")
```
